### Comparing Meme Cryptos with Messari Python Client

Messari's API seems to have the most available endpoints.  The API was a little clumsy to use, but a Python client has been developed with good documentation (https://objective-lalande-8ec88b.netlify.app/index.html).  As with the YFinance notebook, I'm going to use Messari's Python client to compare top meme cryptos (DOGE-related) with market caps over 100m. To date, there are currently seven (7) Doge-meme cryptos (DMCs) with a market cap over 100m USD:

- Dogecoin - DOGE
- SHIBA INU - SHIB
- Dogelon Mars - ELON
- Samoyedcoin - SAMO
- Hoge Finance - HOGE
- Doge Dash - DOGEDASH

YFinance was limited to DOGE and SHIB, so I'll be interested to see what others might be available with the Messari API (note: only DOGE and SHIB are available).  Also, with considerably more endpoints, including supply, the possible for analysis are much greater.

```python
import pandas as pd
from messari.messari import Messari
```

### Getting Time Series Data

There are lots of options for current data and, similarly, for time series data.  First, we will look at price over 2021.

```python
metric = 'price'
start = '2021-05-01'
end = '2021-11-30'
timeseries_df = messari.get_metric_timeseries(asset_slugs=assets, asset_metric=metric, start=start, end=end)
timeseries_df
```

```python
my_list=timeseries_df.columns.values.tolist()
my_list
```

```python
timeseries_df.columns=timeseries_df.columns.map('-'.join).str.strip('-')
```

```python
timeseries_df.head(3)
```

```python
df1=timeseries_df[['dogecoin-close', 'shiba-inu-close']]
df1.head()
```

```python
df1.describe()
```

```python
#Using fivethirtyeight style
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
my_crypto = df1
plt.figure(figsize = (12.2, 4.5))
for c in my_crypto.columns.values:
   plt.plot(my_crypto[c], label = c)
plt.title('DOGE VS SHIB')
plt.xlabel('Days')
plt.ylabel(' Crypto Price ($)')
plt.legend(my_crypto.columns.values, loc= 'upper left')
plt.show()
```

```python
#Scale the data
#the min-max scaler method scales the dataset so that all the input features lie between 0 and 100 inclusive
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0, 100))
scaled = min_max_scaler.fit_transform(df1)
scaled
```

```python
df1_scale = pd.DataFrame(scaled, columns = df1.columns)
```

```python
#Visualize the scaled data
my_crypto = df1_scale
plt.figure(figsize=(12.4, 4.5))
for c in my_crypto.columns.values:
   plt.plot(my_crypto[c], label=c)
plt.title('DOGE VS SHIB (Scaled Graph)')
plt.xlabel('Days')
plt.ylabel('Crypto Scaled Price ($)')
plt.legend(my_crypto.columns.values, loc = 'upper left')
plt.show()
```

```python
DSR = df1.pct_change(1)
DSR
```

```python
plt.figure(figsize=(12, 4.5))
for c in DSR.columns.values:
   plt.plot(DSR.index, DSR[c], label = c, lw = 2, alpha = .7)
plt.title('DOGE VS SHIB - Daily Simple Returns')
plt.ylabel('Percentage (in decimal form')
plt.xlabel('Days')
plt.legend(DSR.columns.values, loc= 'upper right')
plt.show()
```

#### Hashrate
SHIB is not a mineable token, so comparing hashrates is not possible.  But let's look at DOGE and compare hashrates to price.

```python
metric = 'hashrate'
start = '2021-05-01'
end = '2021-12-01'
timeseries_df = messari.get_metric_timeseries(asset_slugs=assets, asset_metric=metric, start=start, end=end)
timeseries_df.head(3)
```

```python
#Quick plot
timeseries_df.plot()
```

```python
timeseries_df.head(3)
```

```python
dogeprice=df1['dogecoin-close']
dogeprice.head(3)
```

```python
merged = timeseries_df.join(dogeprice, how='outer')
merged
```

```python
merged.rename(columns={'dogecoin': 'doge-hashrate'}, inplace=True)
```

```python
#Scale the data
#the min-max scaler method scales the dataset so that all the input features lie between 0 and 100 inclusive
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0, 100))
scaled = min_max_scaler.fit_transform(merged)
scaled
```

```python
merged_scale = pd.DataFrame(scaled, columns = merged.columns)
```

```python
#Visualize the scaled data
my_crypto = merged_scale
plt.figure(figsize=(12.4, 4.5))
for c in my_crypto.columns.values:
   plt.plot(my_crypto[c], label=c)
plt.title('DOGE - Hashrate and Price')
plt.xlabel('Days')
plt.ylabel('Hashrate to Price')
plt.legend(my_crypto.columns.values, loc = 'upper left')
plt.show()
```
